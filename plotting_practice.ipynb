```python
import pandas as pd

df = pd.read_csv('./sample_data/mnist_test.csv')
df.head()
```

```python
# Display basic information about the dataset
print(df.info())

# Display basic statistics about the dataset
print(df.describe())
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Extract the first image and its label
for i in range(100):
  first_image = df.iloc[i-1, i:].values
  first_image = first_image.reshape(28, 28).astype('uint8')
  first_label = df.iloc[i-1, i]

  # Plot the first image
  plt.imshow(first_image, cmap='gray')
  plt.title(f'Label: {first_label}')
  plt.show()
```

```python
# Normalize the pixel values
df.iloc[:, 1:] = df.iloc[:, 1:] / 255.0

# Verify normalization
print(df.head())
```

```python
# Separate features and labels
X = df.iloc[:, 1:].values
y = df.iloc[:, 0].values

# Display the shape of the features and labels
print(f'Features shape: {X.shape}')
print(f'Labels shape: {y.shape}')
```

```python
# Function to plot multiple digits
def plot_digits(images, labels, num_rows=2, num_cols=5):
    fig, axes = plt.subplots(num_rows, num_cols, figsize=(10, 5))
    axes = axes.ravel()
    for i in range(num_rows * num_cols):
        axes[i].imshow(images[i].reshape(28, 28), cmap='gray')
        axes[i].set_title(f'Label: {labels[i]}')
        axes[i].axis('off')
    plt.subplots_adjust(hspace=0.5)
    plt.show()

# Plot the first 10 digits
plot_digits(X, y, num_rows=2, num_cols=5)
```

```python
# Display the distribution of labels
label_counts = df.iloc[:, 0].value_counts()
print(label_counts)

# Plot the distribution of labels
label_counts.plot(kind='bar')
plt.title('Distribution of Labels')
plt.xlabel('Label')
plt.ylabel('Count')
plt.show()
```
